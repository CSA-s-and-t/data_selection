# Insertion of Self Generated videos into dataset

```python
import os
import pandas as pd
import numpy as np

main_df = pd.read_csv('test_old.csv')
main_df
```

```python
main_df = main_df[main_df['augmentation'] != 'statattack']
main_df
```

```python
# insert new augmentations into the dataframe 
augmentation_techniques = ['vmifgsm_3', 'vmifgsm_6', 'decowa_3', 'decowa_6']
augment_df = main_df[main_df['augmentation'] == 'gra_3']
augment_df = augment_df.loc[augment_df.index.repeat(len(augmentation_techniques))].reset_index(drop=True)
augment_df['augmentation'] = np.tile(augmentation_techniques, 10)
augment_df['original_path'] = augment_df['target'].str.removesuffix('.mp4') + '_' + augment_df['augmentation'] + '.mp4'
```

```python
# insert self generated videos into dataframe
path = "D:/datasets/self-generated"
self_generated_videos =  [os.path.join(path, f) for f in os.listdir(path)]

fake = [video for video in self_generated_videos if 'fake' in video or 'Fake' in video]
real = [video for video in self_generated_videos if 'real' in video]

self_fake_df = pd.DataFrame(columns=['original_path', 'label', 'technique', 'source', 'target', 'augmentation', 'tool', 'content'])
self_fake_df['original_path'] = fake
self_fake_df['label'] = 1
self_fake_df['technique'] = 'lipsync-voiceclone'
self_fake_df['source'] = 'SelfGenerated'
self_fake_df['augmentation'] = '-'
self_fake_df['target'] = self_fake_df['original_path'].apply(lambda x: x.replace('fake', 'real').replace('Fake', 'real'))
self_fake_df['tool'] = 'wav2lip-elevenlabs'
self_fake_df.at[0, 'content'] = 'CSA seminar emma watson'
self_fake_df.at[1, 'content'] = 'CSA CSEC josephine teo'
self_fake_df.at[2, 'content'] = 'CSA anniversary lee hsien loong'
self_fake_df.at[3, 'content'] = 'CSA anniversary lee hsien loong'
self_fake_df.at[4, 'content'] = 'CSA smart nation scholarship taylor swift'

augmentation_techniques = ['gra_3', 'gra_6', 'ncs_3', 'ncs_6', 'dem_3', 'dem_6', 'decowa_3', 'decowa_6', 'vmifgsm_3', 'vmifgsm_6']
self_fake_augment_df = self_fake_df.loc[self_fake_df.index.repeat(len(augmentation_techniques))].reset_index(drop=True)
self_fake_augment_df['augmentation'] = np.tile(augmentation_techniques, len(fake))
self_fake_augment_df['original_path'] = self_fake_augment_df['original_path'].str.replace('self-generated', 'augmented')
self_fake_augment_df['original_path'] = self_fake_augment_df['original_path'].str.removesuffix('.mp4') + '_' + self_fake_augment_df['augmentation'] + '.mp4'


self_real_df = pd.DataFrame(columns=['original_path', 'label', 'technique', 'source', 'target', 'augmentation', 'tool', 'content'])
self_real_df['original_path'] = real
self_real_df['label'] = 0
self_real_df['technique'] = '-'
self_real_df['augmentation'] = '-'
self_real_df['source'] = 'SelfGenerated'
self_real_df['target'] = '-'
self_real_df['tool'] = '-'
self_real_df['content'] = '-'
```

```python
#merge and output

main_df = pd.concat([main_df, self_real_df, self_fake_df, self_fake_augment_df], ignore_index=True)
main_df.to_csv('test.csv', index=False)
```
